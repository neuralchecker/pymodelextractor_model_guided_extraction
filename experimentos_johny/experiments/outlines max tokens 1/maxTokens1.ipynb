```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import ks_2samp
```

```python
asmr_data = pd.read_csv('../../csv/asmr/asmr_1k_max_tokens1.csv')
outlines_data = pd.read_csv('../../csv/outlines/outlines_1k_max_tokens1.csv')
llm_data = pd.read_csv('../../csv/llm/llm_1k_max_tokens1.csv')
```

```python
sns.set(style="whitegrid")

fig, axs = plt.subplots(1, 3, figsize=(15, 5), sharey=True)

sns.histplot(outlines_data['floating-point'], bins=10, stat='probability', label='Outlines', ax=axs[0], color = 'blue')
sns.histplot(llm_data['floating-point'], bins=10, stat='probability', label='LLM All Tokens', ax=axs[1], color='green')
sns.histplot(asmr_data['floating-point'], bins=10, stat='probability', label='ASMR All Tokens', ax=axs[2], color='orange')


fig.suptitle('Floating point distribution // Max Tokens = 2 ', fontsize=16)
for ax in axs:
    ax.set_xlabel('Value', fontsize=12)
    ax.set_ylabel('Probability', fontsize=12)
    ax.legend(fontsize=10)

plt.show()
```
