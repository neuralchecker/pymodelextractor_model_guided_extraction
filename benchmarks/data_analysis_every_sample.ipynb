```python
import pandas as pd
```

```python
evaluation_data = pd.read_csv('./results/benchmark_1_1709686830.9642582.csv')
```

```python
evaluation_data.tail(10)
```

```python
evaluation_data['Time'] = evaluation_data['Generation Time'] + evaluation_data['Sample Time']
evaluation_data
```

```python
import seaborn as sns

sns.set_style("whitegrid")
ax = sns.lmplot(x="Samples", y="Time", hue="Algorithm",
                 data=evaluation_data, legend = True, legend_out = False, order = 2, ci=None, height = 6)
ax.axes[0,0].set_xlabel("Samples", fontsize = 12)
ax.axes[0,0].set_ylabel("Time", fontsize = 12)
ax.axes[0,0].set_title("Generation + Sample Time", fontsize = 12)
```

```python
sns.set_style("whitegrid")

ax = sns.lmplot(x="Samples", y="Sample Time", hue="Algorithm",
                 data=evaluation_data, legend = True, legend_out = False, order = 2, ci=None, height = 6)
ax.axes[0,0].set_xlabel("Samples", fontsize = 12)
ax.axes[0,0].set_ylabel("Time", fontsize = 12)
ax.axes[0,0].set_title("Sample Time", fontsize = 12)
```

```python
sns.set_style("whitegrid")

ax = sns.lmplot(x="Samples", y="Generation Time", hue="Algorithm",
                 data=evaluation_data, legend = True, legend_out = False, order = 1, ci=None, height = 6)
ax.axes[0,0].set_xlabel("Samples", fontsize = 12)
ax.axes[0,0].set_ylabel("Time", fontsize = 12)
ax.axes[0,0].set_title("Generation Time", fontsize = 12)
```
