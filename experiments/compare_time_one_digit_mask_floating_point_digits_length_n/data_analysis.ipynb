```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

# Data Loading

```python
evaluation_data = pd.read_csv('results/one_digit_mask_length_1.csv')
evaluation_data.head(10)
```

```python
evaluation_data['Time'] = evaluation_data['Generation Time'] + evaluation_data['Sample Time']
evaluation_data.head(10)
```

```python


sns.set_style("whitegrid")

evaluation_data_filtered = evaluation_data[evaluation_data['Samples'] != 50]


sns.set_style("whitegrid")

plt.figure(figsize=(10, 8))
ax = sns.boxplot(x="Samples", y="Time", hue="Algorithm", data=evaluation_data_filtered, dodge=False)
ax.set_title('Generate + Sample Time')
ax.set_xlabel('Samples')
ax.set_ylabel('Time')

plt.legend(title='Algorithm', loc='upper left', bbox_to_anchor=(0.02, 0.98))

plt.tight_layout()
plt.show()
```

```python
sns.set_style("whitegrid")

plt.figure(figsize=(10, 8))
ax = sns.lineplot(x="Samples", y="Time", hue="Algorithm", data=evaluation_data_filtered, errorbar='sd', marker='o')
ax.set_title('Generate + Sample Time')
ax.set_xlabel('Samples')
ax.set_ylabel('Sample Time')

plt.legend(title='Algorithm', loc='upper left')

plt.tight_layout()
plt.show()
```
