```python
# this is using the floatingPointPartitioner, not the Quantization
asmr_data = pd.read_csv('../../csv/asmr/asmr_10k.csv')
repeated_values_asmr = asmr_data['floating-point'].value_counts()
print(repeated_values_asmr)
```

```python
sns.set(style="whitegrid")

fig, axs = plt.subplots(1, 3, figsize=(20, 7), sharey=True)

sns.histplot(llm_data['floating-point'], bins=20, stat='probability', label='GPT2 ', ax=axs[0], color='green')
sns.histplot(asmr_data_kappa5['floating-point'], bins=20, stat='probability', label='ASMR KAPPA 5', ax=axs[1], color='orange')
sns.histplot(asmr_data['floating-point'], bins=20, stat='probability', label='ASMR', ax=axs[2], color='blue')

fig.suptitle('Floating point distribution', fontsize=16)
for ax in axs:
    ax.set_xlabel('Value', fontsize=12)
    ax.set_ylabel('Probability', fontsize=12)
    ax.legend(fontsize=10)

plt.show()
```
